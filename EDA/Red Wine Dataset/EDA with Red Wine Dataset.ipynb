**Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine**

The key features of the Wine Quality dataset are:

1. **Fixed Acidity**: Influences the wine's crispness and balance.
2. **Volatile Acidity**: Affects taste; higher levels make wine unpleasant.
3. **Citric Acid**: Adds freshness and preserves flavor.
4. **Residual Sugar**: Impacts sweetness; higher levels mean sweeter wine.
5. **Chlorides**: Affects taste; higher levels make wine salty.
6. **Free Sulfur Dioxide**: Preserves wine from spoilage.
7. **Total Sulfur Dioxide**: Affects taste and prevents oxidation.
8. **Density**: Reflects sugar and alcohol content.
9. **pH**: Controls acidity and preservation.
10. **Sulphates**: Preserves wine and enhances taste.
11. **Alcohol**: Impacts the wine's body and quality.

Each feature contributes to predicting the **Quality** score, the target variable in the dataset.

**Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
Discuss the advantages and disadvantages of different imputation techniques.**

If missing data was present, It can be handled by using imputation techniques such as:

1. **Mean/Median Imputation**:
   - *Advantages*: Simple and quick to implement. Preserves the dataset size.
   - *Disadvantages*: Can distort relationships between features, especially with skewed data.

2. **Mode Imputation**:
   - *Advantages*: Useful for categorical features. Maintains the most frequent value.
   - *Disadvantages*: May lead to bias in the data if the mode is overly dominant.

3. **Drop Missing Data**:
   - *Advantages*: Simple, ensures no bias from imputation.
   - *Disadvantages*: Reduces dataset size, potentially leading to information loss.


**Q3. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?**

#### Concept of Normality and Non-Normality:
- **Normality**: A feature follows a bell-shaped, symmetric distribution (normal distribution). It has equal mean, median, and mode, and most of the data lies close to the mean.
- **Non-Normality**: When data does not follow a normal distribution; it may be skewed, have outliers, or a multi-modal distribution.

#### In the Wine Quality dataset:
From the EDA process, the features that exhibit non-normality include:

1. **Residual Sugar**: Right-skewed due to higher values.
2. **Chlorides**: Right-skewed, with a small number of high values.
3. **Density**: Slightly skewed, not normally distributed.

#### Transformations to Improve Normality:
1. **Log Transformation**: Applies to skewed data like Residual Sugar and Chlorides to reduce skewness.
2. **Square Root Transformation**: Reduces the impact of high values in right-skewed data, useful for Density.



**Q4. Using the wine quality data set, perform principal component analysi (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?**


### Principal Component Analysis (PCA)
PCA is a dimensionality reduction technique used to transform a large set of correlated features into a smaller set of uncorrelated components, while retaining as much variance as possible. It helps simplify models and avoid overfitting by reducing the number of features.

### Performing PCA on the Wine Quality dataset:
1. **Standardize the Data**: PCA requires standardized features to ensure equal weight.
2. **Calculate Principal Components**: These are linear combinations of the original features that capture the maximum variance.
3. **Variance Explained**: Determine the number of components needed to capture the desired variance (90%).

In the Wine Quality dataset, performing **PCA** shows that the **first 2 principal components** explain **90%** of the variance.

- **Principal Component 1** explains about 59.87% of the variance.
- **Principal Component 2** explains about 37.89% of the variance.
  
Thus, to reduce dimensionality while retaining 90% of the information, only 2 components are needed.

```python
# Importing necessary libraries for PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# dataset
wine_data = df

# Standardizing the data
scaler = StandardScaler()
scaled_wine_data = scaler.fit_transform(wine_data)

# Performing PCA
pca = PCA()
pca.fit(scaled_wine_data)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Cumulative explained variance to find the number of components that explain 90% variance
cumulative_variance = explained_variance.cumsum()

# Finding the minimum number of components needed to explain 90% variance
components_needed = len([x for x in cumulative_variance if x < 0.90]) + 1

print("Explained Variance Ratio:", explained_variance)
print("Cumulative Variance:", cumulative_variance)
print("Components Needed to explain 90% variance:", components_needed)
```
